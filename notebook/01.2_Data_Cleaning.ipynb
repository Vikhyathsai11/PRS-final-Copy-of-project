# Data Cleaning and Preparation

This notebook prepares the raw traffic data for analysis by handling missing values, removing duplicates, correcting data types, and standardizing important fields. The cleaned dataset is the foundation for all downstream analysis and visualization.

### Step 2: Clean and standardize the dataset
This section prepares the data for analysis by fixing missing values and formatting issues.

```python
# Clean and standardize the traffic dataset
import pandas as pd

# Load cleaned dataset
df = pd.read_csv("../data/cleaned_traffic.csv")

print("Shape before cleaning:", df.shape)
```

```python
# Count missing values
#AI suggestion: Check for missing values and their percentage in the dataset
missing = df.isnull().sum()

# Percentage of missing values
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    "Missing Values": missing,
    "Percentage": missing_percent
})

missing_df = missing_df.sort_values(by="Percentage", ascending=False)

missing_df.head(20)
```

```python
# Columns with 100% missing values
#our idea to remove 100% missing values ,AI help to write syntax to remove columns with 100% missing values
empty_columns = missing_df[missing_df["Percentage"] == 100].index.tolist()

print("Columns to remove:")
print(empty_columns)

# Remove them
df.drop(columns=empty_columns, inplace=True)

print("New Shape:", df.shape)
```

```python
# Columns with more than 99% missing values
#AI suggestion: Identify and remove columns with more than 99% missing values
high_missing = missing_df[missing_df["Percentage"] > 99].index.tolist()

print("Highly missing columns:")
print(high_missing)

# Remove only if still present
df.drop(columns=high_missing, inplace=True, errors='ignore')

print("Shape after removing high missing columns:")
print(df.shape)
```

```python
# Clean and standardize the traffic dataset
removed_columns = [
    "TGMax2_SoFei_Ri1",
    "TGMax2_SoFei_Ri2",
    "Unnamed: 255",
    "Str_Zus"
]

print("Verification:")

for col in removed_columns:
    if col in df.columns:
        print(f"{col} : Still Exists")
    else:
        print(f"{col} : Removed")
```

```python
# Clean and standardize the traffic dataset
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
# Clean and standardize the traffic dataset
df.dtypes.value_counts()
```

```python
#summary statistics for numeric columns
#AI suggestion: Generate summary statistics for numeric columns to understand the distribution and central tendency of the data
missing_after = df.isnull().sum()
missing_after = missing_after[missing_after > 0]

missing_after = missing_after.sort_values(ascending=False)

missing_after.head(20)
```

```python
# Clean and standardize the traffic dataset
df.to_csv("traffic_cleaned_final.csv", index=False)

print("Final dataset saved successfully!")

print("Final Shape:", df.shape)
```
