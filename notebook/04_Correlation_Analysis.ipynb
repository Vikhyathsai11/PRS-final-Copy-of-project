# Correlation and Statistical Analysis

This notebook examines the relationships between traffic volume, heavy vehicle traffic, truck share, and year. The analysis helps uncover whether traffic growth is linked to changes in freight movement or other structural patterns in the dataset.

### Step 5: Measure relationships
This section examines how traffic volume relates to heavy vehicles and year.

```python
# Examine the relationships between traffic variables
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import pearsonr

plt.style.use("ggplot")


pd.set_option("display.max_columns", None)
```

```python
# Examine the relationships between traffic variables
df = pd.read_csv("../data/traffic_cleaned_final.csv")

roads = [6,9,93]

traffic = df[df["Str_Nr"].isin(roads)].copy()

traffic = traffic.dropna(
    subset=[
        "DTV_Kfz_W_Q",
        "DTV_SV_W_Q",
        "pSV_W_Q"
    ]
)

traffic.shape
```

```python
# Examine the relationships between traffic variables
print("="*60)

print("CORRELATION DATASET")

print("="*60)

print("Rows :",len(traffic))

print("Roads :",traffic["Str_Nr"].unique())

print("Years :",traffic["Jahr"].min(),"-",traffic["Jahr"].max())
```

```python
# Examine the relationships between traffic variables
corr_df = traffic[
    [
        "Jahr",
        "Str_Nr",
        "DTV_Kfz_W_Q",
        "DTV_SV_W_Q",
        "pSV_W_Q",
        "Anz_Fs_Q",
        "Betriebs_km"
    ]
]

corr_df.head()
```

```python
# Examine the relationships between traffic variables
corr = corr_df.corr(numeric_only=True)

corr.round(3)
```

```python
# Examine the relationships between traffic variables
print("="*60)
print("CORRELATION ANALYSIS")
print("="*60)

for idx,row in results.iterrows():

    r = row["Correlation"]

    if abs(r)>=0.8:
        strength="Very Strong"

    elif abs(r)>=0.6:
        strength="Strong"

    elif abs(r)>=0.4:
        strength="Moderate"

    elif abs(r)>=0.2:
        strength="Weak"

    else:
        strength="Very Weak"

    print(f"{idx}: {strength} correlation ({r:.3f})")
```

```python
# Examine the relationships between traffic variables
from scipy.stats import linregress
```

```python
# Examine the relationships between traffic variables
plt.figure(figsize=(14,7))

colors = {
    6: "blue",
    9: "red",
    93: "green"
}

for road in [6, 9, 93]:

    road_data = (
        traffic[traffic["Str_Nr"] == road]
        .groupby("Jahr")["DTV_Kfz_W_Q"]
        .mean()
        .reset_index()
    )

    x = road_data["Jahr"]
    y = road_data["DTV_Kfz_W_Q"]

    slope, intercept, r_value, p_value, std_err = linregress(x, y)

    plt.scatter(
        x,
        y,
        color=colors[road],
        alpha=0.7
    )

    plt.plot(
        x,
        intercept + slope*x,
        color=colors[road],
        linewidth=3,
        label=f"A{road} (R²={r_value**2:.3f})"
    )

plt.title("Traffic Trend using Linear Regression")

plt.xlabel("Year")

plt.ylabel("Average Weekday Traffic")

plt.legend()

plt.grid(True)

plt.show()
```

```python
# Examine the relationships between traffic variables
print("="*70)
print("LINEAR REGRESSION RESULTS")
print("="*70)

for road in [6,9,93]:

    road_data = (
        traffic[traffic["Str_Nr"]==road]
        .groupby("Jahr")["DTV_Kfz_W_Q"]
        .mean()
        .reset_index()
    )

    slope, intercept, r, p, std = linregress(
        road_data["Jahr"],
        road_data["DTV_Kfz_W_Q"]
    )

    print(f"\nA{road}")

    print(f"Slope       : {slope:.2f}")

    print(f"R²          : {r**2:.3f}")

    print(f"P-value     : {p:.5f}")

    if p < 0.05:
        print("Trend       : Statistically Significant")
    else:
        print("Trend       : Not Significant")
```

```python
# Examine the relationships between traffic variables
road_results = []

for road in [6,9,93]:

    subset = traffic[
        traffic["Str_Nr"]==road
    ]

    r,p = pearsonr(
        subset["DTV_Kfz_W_Q"],
        subset["DTV_SV_W_Q"]
    )

    road_results.append({
        "Road":f"A{road}",
        "Correlation":r,
        "P-value":p
    })

road_corr = pd.DataFrame(road_results)

road_corr
```

```python
# Examine the relationships between traffic variables
plt.figure(figsize=(8,5))

plt.bar(
    road_corr["Road"],
    road_corr["Correlation"]
)

plt.title("Traffic vs Heavy Vehicle Correlation")

plt.ylabel("Pearson Correlation")

plt.show()
```

```python
# Examine the relationships between traffic variables
for _, row in road_corr.iterrows():

    print("="*40)

    print(row["Road"])

    print("="*40)

    r = row["Correlation"]

    if abs(r)>0.8:

        print("Very Strong Relationship")

    elif abs(r)>0.6:

        print("Strong Relationship")

    elif abs(r)>0.4:

        print("Moderate Relationship")

    else:

        print("Weak Relationship")

    print(f"Correlation : {r:.3f}")

    print(f"P-value : {row['P-value']:.5f}")
```

```python
# Examine the relationships between traffic variables
corr_matrix = corr_df.corr(numeric_only=True)

pairs = []

for i in corr_matrix.columns:

    for j in corr_matrix.columns:

        if i != j:

            value = corr_matrix.loc[i,j]

            if abs(value)>=0.70:

                pairs.append([

                    i,

                    j,

                    value

                ])

network = pd.DataFrame(

    pairs,

    columns=[

        "Variable 1",

        "Variable 2",

        "Correlation"

    ]

)

network = network.drop_duplicates()

network
```

```python
# Examine the relationships between traffic variables
network.sort_values(

    "Correlation",

    ascending=False

)
```
