```python
#German decimal separator is comma, so we need to replace it with a dot before converting to numeric
numeric_columns = [
    "Betriebs_km",
    "Koor_WGS84_N",
    "Koor_WGS84_E"
]

for col in numeric_columns:

    if col in df.columns:

        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", ".", regex=False)
        )

        df[col] = pd.to_numeric(
            df[col],
            errors="coerce"
        )
        
```

```python
#Convert year and road number columns to numeric if they exist
df["Jahr"] = pd.to_numeric(
    df["Jahr"],
    errors="coerce"
)

df["Str_Nr"] = pd.to_numeric(
    df["Str_Nr"],
    errors="coerce"
)
```

```python
#Convert traffic volume columns to numeric
traffic_cols = [
    c for c in df.columns
    if c.startswith("DTV")
]

for col in traffic_cols:

    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
    )

    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )
```

```python
#Convert percentage columns to numeric
percentage_cols = [
    c for c in df.columns
    if c.startswith("p")
]

for col in percentage_cols:

    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
    )

    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )
```
